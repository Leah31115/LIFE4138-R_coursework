### Question 6

Using the `starwars` dataset, identify the individuals with:

- the tallest height
- the lowest mass


HINT: the starwars dataset is part of the dplyr package

```r
# Load in the library tidyverse and data starwars
library(tidyverse)
data(starwars)

# Identifying the individual with the tallest height:
# Pipe starwars to select for the columns name and height. Slice the row with the max height to find the tallest individual.
starwars %>%
  select(name, height) %>%
  slice(which.max(height))
# Yarael Poof has the tallest height of 264


# Identifying the individual with the lowest mass:
# Pipe starwars to select for the columns name and mass. Slice the row with the min mass to find the lightest individual.
starwars %>%
  select(name, mass) %>%
  slice(which.min(mass))
# Ratts Tyerell has the lowest mass of 15
```

### Question 7

The `dplyr` dataset `storms` is part of the US National Oceanic and Atmospheric Administration database of Atlantic hurricanes. It contains multiple repeated measurements from hurricanes. For a given storm, there are measurements from every 6 hours across the life time of the storm.

Use a tidyverse approach with `dplyr` to answer the following:

- how many different storms are contained in the database?
- which year had the most storms?
- which storm, in which year had the highest pressure?

HINT:
unique() will help ensure you remove duplicated data from a vector e.g.
x <- c(1, 1, 2, 2, 3, 3, 4, 4)
unique(x)

```r
library(tidyverse)
data(storms)

# Pipe storms to search for unique names within the name column. Then summarise to find the number of unique storms.
storms %>%
  select(name) %>%
  unique() %>%
  summary()
# This tells you there are 258 unique storms in the database

# Pipe storms to find the year that had the highest number of storms. 
# Group by year, then count the number of storm entries (counts all storm rows thus gets multiple counts for unique storms since it will count for day and hours)
# Arrange the years to show the highest count at the top
storm_count <- storms %>%
  group_by(year) %>%
  count() %>%
  arrange(desc(n))

# Year with the most storms
storm_count[1,]
# The year 2005 had the most storms with a total count of 873 storms.


# Pipe storms to select for the name and pressure columns. Then slice the row which has the highest pressure value.
storms %>%
  select(name, pressure) %>%
  slice(which.max(pressure))
# Storm Danny has the highest pressure value of 1024.
```

### Question 8

The `ggplot2` dataset `diamonds` contains the price and information from the sale of 54,000 different diamonds. You can use ?diamonds to find out more about the dataset if you want.

Using `ggplot2`:

- plot the relationship between the carat (i.e. the weight of the diamond) and the price.
- plot a histogram of prices of all the diamonds
- plot a boxplot of the difference in price between the different cuts

```r
library(ggplot2)
data(diamonds)

# A scatterplot to show the relationship between diamond carat and price.
ggplot(diamonds, aes(x = carat, y = price)) +
    geom_point(colour = "blue", alpha = 0.25) +
    theme_light() +
    ggtitle("Diamond carat vs price") +
    xlab("Carat") +
    ylab("Price ($)") +
    theme(plot.title = element_text(hjust = 0.5))
  
# A histogram of diamond price
ggplot(diamonds, aes(x=price)) +
    geom_histogram(binwidth = 250, colour = "black", fill = "orange") + 
    xlab("Price ($)") +
    ggtitle("Histogram of diamond price") +
    theme(plot.title = element_text(hjust = 0.5))

# A boxplot of the difference between price between the different cuts
ggplot(diamonds, aes(x = cut, y = price)) +
    geom_boxplot(colour = "black", fill = "lightgrey") +
    theme_light() +
    ylab("Price ($)") +
    xlab("Diamond cut quality") +
    ggtitle("Prices of diamond cuts") +
    theme(plot.title = element_text(hjust = 0.5))
```

### Question 9

Identify the maximum and minimum price that the diamonds in the diamonds data set sold for. Add a new column to the diamonds data set that contains the string “Low price” if the value in the price column is less than 5000 dollars and “High price” if it is above 5000 dollars.

```r
# Max. diamond price
print("Max diamond cost")
which.max(diamonds$price)

# Min. diamond price
print("Min diamond cost")
which.min(diamonds$price)

# Create new column called price_category in diamonds
diamonds$price_category <- NA

# Since the question doesn't state what category a diamond with price == 5000 should be, I have assumed 5000 to be included within the "High price" category
# Assign "High price" to the diamonds that are greater or equal to 5000, else assign "Low price" to any diamonds that are less than 5000 in price.
diamonds$price_category <- ifelse(diamonds$price >= 5000, "High price", "Low price")
diamonds
```

### Question 10

Sticking with the `diamonds` dataset for this question. A carat is a special unit of weight for gemstones where a single carat is 200 mg or 0.2 g. Write a function to convert the `diamonds$carat` vector into grams or milligrams. How heavy is the heaviest diamond?

```r
# Function to convert carats to mg
carat_to_mg_converter <- function(carat){
    mass_mg <- carat * 200
    
    return(mass_mg)
}

# Convert the carat into mgs with the values in a new column called mass_mg
diamonds["mass_mg"] <- lapply(diamonds["carat"], carat_to_mg_converter)

diamonds

# Find heaviest diamond
diamonds %>% 
    slice(which.max(mass_mg))
# The heavist diamond weighs 1002 mg
```
